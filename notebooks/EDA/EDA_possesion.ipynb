# Analysis by team (aggregated across multiple seasons)

```python
num_cols_final = [
    "possession",
    "take_ons_won_pct",
    "take_ons_tackled_pct",
    "touches_per90",
    "progressive_carries_per90",
    "dispossessed_per90",
]
```

```python
team_summary = (
    df
    .groupby("team")[num_cols_final]
    .agg(["mean", "median", "min", "max"])
    .round(3)
)

team_summary.T
```

### 1) Team-level differences (multi-season view)
Aggregating by team across seasons, possession metrics show **clear and persistent differences** between Premier League clubs. Some teams consistently dominate the ball, while others deliberately concede possession and prioritize defensive or counter-attacking approaches. This suggests possession reflects a teamâ€™s **long-term tactical identity**, not just seasonal fluctuation.

### 2) High-possession teams
Teams such as **Manchester City, Arsenal, Chelsea, and Tottenham** stand out with high average possession, high touches_per90, and strong progressive_carries_per90. **Manchester City** leads across all three, indicating comprehensive control and build-up ability. Their relatively tight median and minâ€“max ranges also point to a stable style across seasons.

### 3) Mid-possession teams
Teams like **Aston Villa, Manchester United, Newcastle United, Fulham, and Brighton** sit in the middle: moderate possession and touches, but still solid ball progression (progressive_carries_per90). This profile suggests a hybrid approachâ€”**selective control** combined with faster transitions rather than full dominance.

### 4) Low-possession teams
On the other end, **Sheffield United, Burnley, Crystal Palace, Everton, and Sunderland** show lower possession alongside limited touches_per90 and progressive_carries_per90. This aligns with low-block defending and more direct counter-attacking. Some teams here still maintain reasonable take_ons_won_pct, highlighting the importance of individual actions within a lower-possession style.

### 5) Control vs risk
The pair of take_ons_won_pct and take_ons_tackled_pct highlights a clear **efficiencyâ€“risk trade-off**. High-possession teams do not always have low tackled rates, and dispossessed_per90 can remain high for top sides simply because they handle the ball more oftenâ€”indicating higher volume-driven risk rather than weaker skill.

### 6) Consolidated conclusion
Overall, possession is a **structural team characteristic** that remains stable across seasons. Combining possession share, touch intensity, ball progression, and turnover risk helps separate teams by playing styleâ€”from dominant control to defensive counter-attackingâ€”providing a strong base for next steps such as style clustering or incorporating possession features into machine-learning models.


```python
team_scatter = (
    df
    .groupby("team")
    .agg(
        possession_mean=("possession", "mean"),
        progressive_carries_per90_mean=("progressive_carries_per90", "mean")
    )
    .reset_index()
)
```

To analyze each teamâ€™s stable possession profile, the data is aggregated at the **team level** rather than by season. This reduces noise from short-term seasonal fluctuations and better reflects each clubâ€™s long-term playing style.

**Two aggregated metrics:**
- `possession`: overall possession share (quantity of possession)
- `progressive_carries_per90`: ball progression via progressive carries (quality of possession)

For each team, we use the **mean** of these metrics across multiple seasons to represent its typical level. This enables direct team-to-team comparison and supports a scatter plot to examine the relationship between possession quantity and possession quality.


```python
plt.figure(figsize=(12, 8))

plt.scatter(
    team_scatter["possession_mean"],
    team_scatter["progressive_carries_per90_mean"],
    alpha=0.75,
    edgecolors="black"
)

# Label each team
for _, row in team_scatter.iterrows():
    plt.text(
        row["possession_mean"] + 0.1,
        row["progressive_carries_per90_mean"] + 0.05,
        row["team"],
        fontsize=9
    )

plt.xlabel("Average possession share (%) (possession)")
plt.ylabel("Progressive carries per 90 (progressive_carries_per90)")
plt.title("Relationship between possession quantity and possession quality by team")

plt.grid(alpha=0.3)
plt.show()
```

## 1) Overall pattern
The chart shows a **clear positive relationship** between possession share and progressive carries per 90. In general, teams that keep the ball more also progress it moreâ€”but the increase is **not perfectly linear**, reflecting different ways teams use possession.

## 2) High possession + strong progression (upper-right)
Teams in the upper-rightâ€”**Manchester City, Liverpool, Arsenal, Chelsea, Tottenham**â€”combine high possession with high progressive carries, indicating efficient use of the ball to control games and move play forward. **Manchester City** stands out as the clearest top performer.

## 3) Higher possession, moderate progression
Teams such as **Manchester United, Brighton, Fulham** show relatively high possession but only mid-level progressive carries. This suggests a more **tempo-control** approachâ€”keeping the ball without constantly driving forward via carries.

## 4) Lower possession, relatively strong progression
Teams like **Bournemouth, Luton Town, Nottingham Forest** have lower possession yet relatively good progressive carries, hinting at a more **direct style**â€”using quick carries during shorter possession spells.

## 5) Low possession + weak progression (lower-left)
**Sheffield United, Sunderland, Crystal Palace, Everton** sit in the lower-left, with both low possession and limited progressive carryingâ€”consistent with deeper defending, simpler counter-attacks, or squad-quality constraints.

## 6) Notable cases
- **Manchester City**: the most prominent outlier, elite in both possession and progression.
- **Sheffield United**: a low-end outlier, limited in both ball control and progression.
- **Brighton**: relatively efficient progression despite not being among the highest-possession teams.

## 7) Consolidated conclusion
Possession is most meaningful when paired with **effective progression**. The strongest teams do not just keep the ballâ€”they convert possession into forward movement, while low-control/low-progression teams struggle to impose themselves, leading to clear stylistic separation across the league.
