```python
num_cols = [
    # passing quality
    "passes_completion_pct",
    "passes_completion_pct_short",
    "passes_completion_pct_medium",
    "passes_completion_pct_long",

    # creativity & efficiency
    "xag",
    "assists_minus_xag"
]
```

**`passes_completion_pct`**

* Captures a teamâ€™s **overall pass completion rate**, serving as a proxy for baseline passing quality and technical security.
* As a **rate-based metric**, it is not driven by the number of matches played, making it suitable for season-to-season comparisons.


**`passes_completion_pct_short`**

* Measures **short-pass accuracy**, a fundamental skill for sustaining possession and controlling tempo.
* Typically stable, helping identify how **consistent** teams are in core technical execution across the league.


**`passes_completion_pct_medium`**

* Reflects effectiveness in **medium-range passing**, where risk and technical demands are higher than for short passes.
* Highlights **tactical differences** in how teams progress the ball and build attacks.


**`passes_completion_pct_long`**

* Quantifies the success rate of **long passing**, a higher-variance and higher-risk action that can produce rapid progression and create breakthroughs.
* Often shows greater dispersion, enabling clearer differentiation between teams.


**`xag`**

* Represents the **quality of chances created from passes**, and is generally less sensitive to finishing luck than raw assists.
* Enables a more objective assessment of **attacking creativity** than actual assist totals alone.


**`assists_minus_xag`**

* Captures **finishing/conversion performance relative to expectation**.
* Helps distinguish teams that create high-quality chances but finish poorly from teams that convert chances above expectation.

# xag_per90

**Definition:** `xag_per90` measures the quality of chances created from passes per 90 minutes. By normalizing for playing time, it removes differences caused by uneven match totals across seasons and enables fair comparisons of teamsâ€™ chance-creation ability.


```python
df["xag_per90"] = df["xag"] / df["minutes_90s"]
```

```python
plt.figure(figsize=(6, 4))
df.boxplot(column="xag_per90", by="season")
plt.title("xag by Season")
plt.suptitle("")
plt.xlabel("Season")
plt.ylabel("xag")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

* **2023â€“2024**: The median is high and the IQR is relatively wide, indicating strong overall chance creation per 90 but with noticeable differences in creativity across teams.

* **2024â€“2025**: The median is similar to (or slightly higher than) the previous season, while dispersion narrows, suggesting more stable and more uniform chance creation from passing.

* **2025â€“2026**: The median is lower than the other two seasons, with a few high-end outliers. This points to a lower league-wide creative baseline per 90, while a small number of teams still generate chances at an exceptional level.


# assists_minus_xag_per90

**Definition:** `assists_minus_xag_per90` measures chance conversion efficiency relative to expectation per 90 minutes. Positive values indicate finishing/conversion above expectation, while negative values suggest under-conversion (wasted chances), providing a deeper view of teamsâ€™ attacking efficiency.


```python
df["assists_minus_xag_per90"] = df["assists_minus_xag"] / df["minutes_90s"]
```

```python
plt.figure(figsize=(6, 4))
df.boxplot(column="assists_minus_xag_per90", by="season")
plt.title("assists_minus_xag_per90 by Season")
plt.suptitle("")
plt.xlabel("Season")
plt.ylabel("assists_minus_xag_per90")
plt.grid(axis="y", alpha=0.3)
plt.show()
```

* **2023â€“2024**: The median is slightly positive, suggesting that teams, on average, convert chances better than expected per 90. However, the IQR is relatively wide and there is a deep negative outlier, indicating that some teams still under-convert despite creating decent chance quality.

* **2024â€“2025**: The median is close to the previous season and the IQR narrows, implying more stable chance conversion per 90 and less variation between teams.

* **2025â€“2026**: The median is near zero or slightly negative, with a wider IQR, indicating weaker chance conversion per 90 and larger differences across teamsâ€”partly influenced by the season being incomplete.

# Analysis by team (aggregated across multiple seasons)

- Teams such as **Tottenham, Brighton, Nottâ€™ham Forest** show a relatively wide **maxâ€“min range**, indicating that passing performance fluctuates notably by season.

## 2) Passing by distance (short / medium / long)

- For **short and medium passing**, top teams (Man City, Chelsea, Arsenal, Liverpool, Brighton) post very high means (â‰ˆ 89â€“93% short, â‰ˆ 88â€“91% medium), reflecting strong technical consistency among possession-oriented sides.
- **Long passing (passes_completion_pct_long)** is the strongest differentiator:
  - Man City, Brighton, and Chelsea maintain means around 54â€“61%.
  - Several bottom teams such as **Sheffield Utd, Burnley, Luton Town** are closer to 40â€“45%.
- The minâ€“max spread for long passing is clearly larger than for short/medium, confirming that long passes are higher-risk and more tactic-dependent.

## 3) Chance creation (xag_per90)

- **Manchester City, Liverpool, Brighton, Chelsea, Arsenal** lead in `xag_per90_mean`, reflecting stable ability to generate high-quality chances per 90 through passing.
- Lower teams such as **Sheffield Utd, Luton Town, Burnley, Sunderland** have low `xag_per90_mean` (< 0.8), indicating limited ability to create dangerous chances from passing.
- Teams like **Tottenham, Newcastle, Aston Villa** have `max` values that are notably higher than their `mean`, suggesting occasional high-ceiling seasons but less consistency.

## 4) Chance conversion vs expectation (assists_minus_xag_per90)

- **Arsenal, Tottenham, Aston Villa, Manchester City** show clearly positive `assists_minus_xag_per90_mean`, indicating above-expectation conversion per 90.
- Many teams such as **Everton, Crystal Palace, Manchester Utd, Brighton, Chelsea** have means near zero or negative, suggesting finishing output does not fully match chance quality.
- The minâ€“max range is wide for many teams (e.g., Everton, Tottenham), implying strong season-to-season swings that depend heavily on attacking personnel.

## 5) Overall summary

- Top teams tend to combine **high passes_completion_pct + high xag_per90**, highlighting the link between possession security and chance creation.
- **Long-pass accuracy and chance conversion** are the two biggest drivers of separation between teams, both in level and in stability.
- Using multi-season **meanâ€“minâ€“max** captures not only average performance but also stability and ceiling, which is useful for next steps such as ranking, clustering, or multivariate analysis.


```python
plt.figure(figsize=(8, 6))

plt.scatter(
    team_summary[('passes_completion_pct', 'mean')],
    team_summary[('xag_per90', 'mean')],
    alpha=0.8,
    edgecolors="black"
)

for team in team_summary.index:
    plt.text(
        team_summary.loc[team, ('passes_completion_pct', 'mean')] + 0.02,
        team_summary.loc[team, ('xag_per90', 'mean')] + 0.005,
        team,
        fontsize=8
    )

plt.xlabel("Pass completion rate (%) (passes_completion_pct)")
plt.ylabel("xG assisted per 90 (xag_per90)")
plt.title("Passing quality vs. chance creation by team")

plt.grid(alpha=0.3)
plt.show()
```

## 1) Overall pattern
- The plot shows a **clear positive association** between average pass completion and average xAG per 90: teams that pass more accurately tend to create more dangerous chances.
- However, the relationship is **not perfectly linear**, suggesting that tactics and personnel also play a role beyond pure passing quality.

## 2) Top-performing cluster (upper-right)
- **Manchester City** stands out in the upper-right, indicating elite performance in both passing control and chance creation quality.
- **Liverpool, Chelsea, Arsenal** also sit in this zone, showing consistently high pass accuracy alongside strong xAG per 90.
- This cluster represents a **control-based style with effective attacking output**.

## 3) Good passing, moderate creativity
- **Brighton, Aston Villa, Manchester Utd, Tottenham, Newcastle Utd** maintain relatively high pass completion, but their xAG per 90 is only mid-to-high.
- This suggests solid ball control, but **less efficient translation into high-quality chances** compared to the top cluster.

## 4) Moderate creativity with lower passing quality
- **Crystal Palace, Brentford, Bournemouth** achieve relatively steady xAG per 90 despite lower pass completion.
- This hints at a greater reliance on **direct play or transitions** rather than sustained possession.

## 5) Weak on both dimensions (lower-left)
- **Sheffield Utd, Luton Town, Burnley, Sunderland, Ipswich Town** sit in the low-pass / low-xAG region.
- This group shows clear limitations in both ball retention and chance creation, reflecting a notable gap versus top teams.

## 6) Notable cases
- **Southampton** shows relatively high pass completion but low xAG per 90, suggesting a possession-oriented approach that lacks end-product in creating dangerous chances.
- **Fulham, West Ham, Wolves** cluster near the center, representing more balanced but not standout performance on either axis.

## 7) Consolidated conclusion
- The scatter plot confirms that **passing quality is an important foundation for chance creation**, but **not a sufficient condition**.
- Top teams combine accurate passing with the ability to **turn control into high-quality chances**, while mid- and lower-table teams often lack one of these components.
- This provides a solid basis for next steps such as **style-based clustering** or **deeper multivariate analysis**.
